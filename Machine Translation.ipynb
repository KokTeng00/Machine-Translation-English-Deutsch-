##### Data Collected from: https://www.kaggle.com/datasets/aditya01233/german-english-neural-machine-translation-dataset

```python
# import the necessary libraries and load the data into a pandas dataframe
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

text_df = pd.read_csv("GERMAN_ENGLISH_TRANSLATION.csv")
```

```python
# display the first 5 rows of the dataframe
text_df.head()
```

```python
# compute the length of the sentences in both languages and plot the distribution
text_df["english_sentence_length"] = text_df["ENGLISH"].apply(lambda x : len(x.split()))
text_df["german_sentence_length"] = text_df["GERMAN"].apply(lambda x : len(x.split()))

plt.figure(figsize = (10, 5))
sns.histplot(text_df["english_sentence_length"], kde = True, color = "blue", label = "English")
sns.histplot(text_df["german_sentence_length"], kde = True, color = "red", label = "German")
plt.legend(title = "Language")
plt.title("Sentence Length Distribution")
plt.xlabel("Sentence Length")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
